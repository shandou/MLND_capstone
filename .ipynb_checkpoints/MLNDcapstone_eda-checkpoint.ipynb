## Extract a random sample 1% of the original size

```python
%%time
raw_csv = './data/train.csv'
sample_csv = './data/train_sample.csv'
need_to_sample = False

if need_to_sample:
    result = subprocess.check_output('wc -l {}'.format(raw_csv), shell=True)
    nlines_total = int(result.split()[0])
    nlines_sample = int(0.01 * nlines_total)
    subprocess.call('head -1 {} > {}'.format(raw_csv, sample_csv), shell=True)
    subprocess.call('gshuf -n {} {} >> {}'.format(nlines_sample, raw_csv, sample_csv), shell=True)
```

```python
# Load sampled csv into pandas dataframe, sort by click time (and reset index)
df = pd.read_csv(
    sample_csv, parse_dates=['click_time', 'attributed_time']
).sort_values(by='click_time').reset_index()
```

```python
df.head()
```

```python
df.info()
```

```python
for col in ['click_time', 'attributed_time']:
    df[col] = pd.DatetimeIndex(df[col]).tz_localize('utc').tz_convert('Asia/Shanghai')
```

```python
df['delta_time_click_vs_attr'] = (df['attributed_time'] - df['click_time']).dt.total_seconds()
df['delta_time_click_vs_attr'].describe()
```

```python
df['click_hour'] = df['click_time'].dt.hour
df['attr_hour'] = df['attributed_time'].dt.hour
df.head()
```

```python
ft2check = list(
    set(df.columns) ^ set(['click_time', 'attributed_time', 'is_attributed', 'index'])
)
```

```python
df_unique = pd.DataFrame(df[ft2check].nunique()).T
df_unique
```

```python
hist_params = dict(bins=24, edgecolor='w', alpha=0.5)
no_params = dict(facecolor='lightcoral', label='No download')
yes_params = dict(facecolor='green', label='Yes download')
fig, ax = plt.subplots()
metric = 'click_hour'
ax.hist(df[df['is_attributed'] == 0][metric], **hist_params, **no_params)
ax.hist(df[df['is_attributed'] == 1][metric], **hist_params, **yes_params)
ax.legend(loc=4)
ax.set(
    yscale='log', xlabel='Click time in hour-of-the day (Shanghai time)', ylabel='Count'
)
plt.show();
```

```python
hist_params = dict(bins=24, edgecolor='w', alpha=0.5)
no_params = dict(facecolor='lightcoral', label='No download')
yes_params = dict(facecolor='green', label='Yes download')
fig, ax = plt.subplots()
metric = 'ip'
ax.hist(df[df['is_attributed'] == 0][metric], **hist_params, **no_params)
ax.hist(df[df['is_attributed'] == 1][metric], **hist_params, **yes_params)
ax.legend(loc=4)
ax.set(
    yscale='log', xlabel='IP groups (encoded)', ylabel='Count'
)
plt.show();
```

```python
attr_ref = 'ip'
attr_list = ['click_hour', 'os', 'device', 'channel', 'app']
```

```python
import itertools
from pprint import pprint
ft_combined = []
for i in range(1, len(attr_list)+1):
    bla.append(list(itertools.combinations(attr_list, i)))
combination_list = list(itertools.chain(*bla))
pprint(combination_list)
```

```python
df_engineered = df.copy()
for attr_other in combination_list:
    print(attr_other)
    tag = '-'.join(attr_other)
    print(tag)
    group_metric = [attr_ref] + list(attr_other)
    df_temp = df.groupby(group_metric).count()['is_attributed']
    df_temp = df_temp.reset_index().rename(columns={'is_attributed' : 'count_ip-{}'.format(tag)})
    df_summary = df.merge(df_temp, how='outer', on=group_metric)
    df_engineered['count_ip-{}'.format(tag)] = df_summary['count_ip-{}'.format(tag)]
```

```python
sorted_counts = sorted([x for x in df_engineered.columns if 'count' in x], key=len)
print(['{}: {}'.format(i, x) for i, x in enumerate(sorted_counts)])
```
